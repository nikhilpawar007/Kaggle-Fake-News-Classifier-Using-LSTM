```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/Fake News Classifier/train.csv')
df.tail()
```

```python
df = df.dropna()
```

```python
x = df.drop('label', axis=1)
x.head()
```

```python
x.shape
```

```python
y = df['label']
y.shape
```

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM,Dense
```

```python
voc_size = 5000
```

```python
messages = x.copy()
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus=[]
for i in range(0, len(messages)):
  print(i)
  review = re.sub('[^a-zA-Z]',' ', messages['title'][i])
  review= review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
one_hot_rep = [one_hot(word,voc_size) for word in corpus]
one_hot_rep
```

```python
sent_length = 20
embedded_docs = pad_sequences(one_hot_rep, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
len(embedded_docs)
```

```python
embedded_docs[0]
```

```python
embedding_vector_feature = 40
model= Sequential()
model.add(Embedding(voc_size, embedding_vector_feature, input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
x_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
x_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_final, y_final, test_size= 0.33, random_state=42)
```

```python
x_train.shape
```

```python
model.fit(x_train,y_train, validation_data=(x_test,y_test), epochs = 10, batch_size= 64)
```

```python
y_pred = model.predict_classes(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
